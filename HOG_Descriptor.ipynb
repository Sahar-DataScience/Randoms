Hog testé sur im1 et double résultats(paramétres modifiés)


```python
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import data, exposure
%matplotlib inline
path = '/content/sample_data/im1.png'
image = plt.imread(path)

fd, hog_image1 = hog(image, orientations=5, pixels_per_cell=(5, 5),
                    cells_per_block=(2, 2), visualize=True, multichannel=True)
fd, hog_image2 = hog(image, orientations=8, pixels_per_cell=(16, 16),
                    cells_per_block=(1, 1), visualize=True, multichannel=True)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 15), sharex=True, sharey=True)


#ax1.axis('off')
#ax1.imshow(image, cmap=plt.cm.gray)
#ax1.set_title('Input image')

# Rescale histogram for better display
hog_image_rescaled1 = exposure.rescale_intensity(hog_image1, in_range=(0, 5))
hog_image_rescaled2 = exposure.rescale_intensity(hog_image2, in_range=(0, 10))

ax1.axis('off')
ax1.imshow(hog_image_rescaled1, cmap=plt.cm.gray)
ax1.set_title('HOG 5')


ax2.axis('off')
ax2.imshow(hog_image_rescaled2, cmap=plt.cm.gray)
ax2.set_title('Histogram of Oriented Gradients 8')
plt.show()
```
